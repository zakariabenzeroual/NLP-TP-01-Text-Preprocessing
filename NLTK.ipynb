```python
from codecs import utf_16_be_decode
import nltk, string

```

#### 1- Loading data :

```python
text_AR = u'ربما كانت أحد أهم التطورات التي قامت بها الرياضيات العربية التي بدأت في هذا الوقت بعمل الخوارزمي و هي بدايات الجبر، و من المهم فهم كيف كانت هذه الفكرة الجديدة مهمة، فقد كانت خطوة ثورية بعيدا عن المفهوم اليوناني للرياضيات التي هي في جوهرها هندسة، الجبر كان نظرية موحدة تتيح الأعداد الكسرية و الأعداد اللا كسرية، و المقادير الهندسية و غيرها، أن تتعامل على أنها أجسام جبرية، و أعطت الرياضيات ككل مسارا جديدا للتطور بمفهوم أوسع بكثير من الذي كان موجودا من قبل، و قدم وسيلة للتنمية في هذا الموضوع مستقبلا و جانب آخر مهم لإدخال أفكار الجبر و هو أنه سمح بتطبيق الرياضيات على نفسها بطريقة لم تحدث من قبل'

text_EN ="Perhaps one of the most significant advances made by Arabic mathematics began at this time with the work of al-Khwarizmi, namely the beginnings of algebra. It is important to understand just how significant this new idea was. It was a revolutionary move away from the Greek concept of mathematics which was essentially geometry. Algebra was a unifying theory which allowed rational numbers, irrational numbers, geometrical magnitudes, etc., to all be treated as \"algebraic objects\". It gave mathematics a whole new development path so much broader in concept to that which had existed before, and provided a vehicle for future development of the subject. Another important aspect of the introduction of algebraic ideas was that it allowed mathematics to be applied to itself in a way which had not happened before."

print(text_EN)
print(text_AR)
```

#### 2 - Convert text to lowercase, Remove punctuation, Word Tokenization and Sentence Tokenization:

**Call nltk.RegexpTokenizer(pattern) with pattern as r"\w+" to create a tokenzier that uses pattern to split a string. 
Call RegexpTokenizer.tokenize(text) with RegexpTokenizer as the previous result and text as a string representing a sentence to return text as a list of words with punctuation's removed.**


```python
#Convert text to lowercase :

text_EN =text_EN.lower()

# Remove Punctuation :

tokenizer = nltk.RegexpTokenizer(r"\w+")

# Word Tokenizer :

words_EN = tokenizer.tokenize(text_EN)
print(words_EN)

words_AR = tokenizer.tokenize(text_AR)
print(words_AR)

# Sentence Tokenizer :

from nltk.tokenize import sent_tokenize

sent_EN = sent_tokenize(text_EN)
print(sent_EN)

sent_AR = sent_tokenize(text_AR)
print(sent_AR)
```

#### 3- Stemming and lemmatization : 

**3-1 Stemming :**

```python
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()

#Word EN :

stemmed = [porter.stem(word) for word in words_EN]
print(stemmed)

#Word AR :

stemmed = [porter.stem(word) for word in words_AR]
print(stemmed)
```

**3-2 lemmatization :**

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

# word EN :

for word in words_EN:
    print(lemmatizer.lemmatize(word))

# word AR :

for word in words_AR:
    print(lemmatizer.lemmatize(word))
```

#### 4- POS Tagger :

```python
from nltk import pos_tag

# Pos Tagger EN : 

pos_EN = pos_tag(words_EN)
print(pos_EN)

# Pos Tagger AR : 

pos_AR = pos_tag(words_AR)
print(pos_AR)
```

#### 5- chunking : 

```python
from nltk import ne_chunk

# chunking EN :

chunks_EN = nltk.ne_chunk(pos_EN)
for tree in chunks_EN:
    print(tree)

# chunking AR :

chunks_AR = nltk.ne_chunk(pos_AR)
for tree in chunks_AR:
    print(tree)
```

#### 6- Parsing :

```python
grammar = "NP: {<DT>?<JJ><NN|NNS>+}"
chunk_parser = nltk.RegexpParser(grammar)

# Parsing EN :

tree_EN =chunk_parser.parse(pos_EN)
tree_EN.draw()

# Parsing AR :

tree_AR =chunk_parser.parse(pos_AR)
tree_AR.draw()
```
